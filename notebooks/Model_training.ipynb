```python
import pandas as pd
data=pd.read_csv("adult.data",names = ['age', 'job','fnlwgt','Qualification','education-num','Marital-Status','occupation','relationship','Colour','Gender','capital-gain','capital-loss','hours-per-week','City','Salary'])

df = data.sample(frac=0.25, random_state=42)
df.head()
```

```python
df.info()
```

```python
def whitespace_remover(dataframe):

  for i in dataframe.columns:

    if dataframe[i].dtype == 'object':

      dataframe[i] = dataframe[i].map(str.strip)

    else:

      pass

whitespace_remover(df)
```

```python
for i in df.columns:
  print(df[i].unique())
```

```python
import numpy as np
df.replace('?',np.nan,inplace =True)
```

```python
df.drop(labels=['fnlwgt','capital-loss'],axis=1,inplace=True)
```

```python
income_map = {'<=50K':0,'>50K':1}

df['Salary'] = df['Salary'].map(income_map)
```

```python
df
```

```python
df.shape
```

```python
X = df.iloc[:, :-1]
```

```python
X.columns
```

```python
y = df['Salary']
```

```python
y.shape
```

```python
numerical_feature = X.select_dtypes(include = 'int64').columns
categorical_feature = X.select_dtypes(include = 'object').columns
```

```python
print(f'numerical columns are : {numerical_feature}')
print(f'categorical columns are : {categorical_feature}')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline(
    steps = [
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]
)

cat_pipeline = Pipeline(
    steps = [
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder',  OneHotEncoder(sparse_output=False)),
        ('scaler', StandardScaler())
    ]
)

preprocessor = ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_feature),
    ('cat_pipeline,',cat_pipeline,categorical_feature)
])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
```

```python
X_test = pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
parameter = {'penalty':['l1', 'l2', 'elasticnet', 'None'],'C':[1,2,0.1,0.001,0.05,0.5],'max_iter':[100,200,300]}
```

```python
regressor = LogisticRegression()
```

```python
regressor_classifier = GridSearchCV(regressor,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
regressor_classifier.fit(X_train,y_train)
```

```python
print(regressor_classifier.best_params_)
```

```python
print(regressor_classifier.best_score_)
```

```python
y_pred = regressor_classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
